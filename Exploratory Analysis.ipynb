```python
import numpy as np
import pandas as pd
import os, json, math, time
import matplotlib.pyplot as plt
import seaborn as sns
```

Concatenate the dataframes

```python
df_1 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_1.head()
```

```python
df_2 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df_2.head()
```

```python
dataframes = [df_1, df_2]
df_merge = pd.concat(dataframes)
display(df_merge.head())
```

```python
df_merge.to_csv("Data/tmdb_results_combined.csv.gz",compression='gzip',index=False)
```

```python
df=df_merge
```

```python
df_plot = df[(df['budget'] > 0) & (df['revenue'] > 0)]
```

```python
print(f"{df_plot.shape[0]} movies had some valid financial information.")
```

```python
print('- certification:')
print(df_plot['certification'].value_counts(dropna=False))
```

```python
avg_revenue = df_plot.groupby('certification')['revenue'].mean().reset_index()
avg_revenue.sort_values('revenue', ascending=False)
```

Kids movies are the most profitable! who knew!

```python
avg_budget = df_plot.groupby('certification')['budget'].mean().reset_index()
avg_budget.sort_values('budget', ascending=False)
```
